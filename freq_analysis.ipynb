```python
import plotly.express as px
import pandas as pd
import numpy as np

filename = 'spectrum2.txt'
```

```python
data = {'frequency': [], 'level': []}

with open(filename, 'rb') as f:   
    i = 0
    for line in f:
        line = line.decode(errors='ignore')
        measurement = line.strip().split('\t')
        if i != 0:
            data['frequency'].append(float(measurement[0]))
            data['level'].append(float(measurement[1]))
        i += 1


df = pd.DataFrame(data)
df
```

```python
import plotly.graph_objects as go
from math import log

fig = go.Figure()

trace1 = go.Scatter(
    x = df['frequency'],
    y = df['level'],
    mode = 'lines',
    name = 'data'
)

layout = dict(
    title = 'Frequency analysis(0s)',
    xaxis = dict(title='frequency [Hz]'),
    yaxis = dict(title='level [dB]'),
)

fig = go.Figure(
    data = [trace1],
    layout = layout
)

fig.add_annotation(
    x=log(28.56445) / log(10), 
    y=-82.60582,
    text='1st'
)
fig.add_annotation(
    x=log(57.86133) / log(10), 
    y=-86.78249,
    text='2nd'
)
fig.add_annotation(
    x=log(169.1895) / log(10), 
    y=-76.6216,
    text='6th'
)
fig.add_annotation(
    x=log(276.8555) / log(10), 
    y=-80.61135,
    text='10th'
)

fig.update_xaxes(type='log')

fig.show()
```
